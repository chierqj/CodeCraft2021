```python
class Server:
    def __init__(self, name, cpu, ram, hv, ev):
        self.name = name
        self.cpu = cpu
        self.ram = ram
        self.hv = hv
        self.ev = ev
    def debug(self):
        print("name: {}, cpu: {}, ram: {}, hv: {}, ev: {}".format(self.name, 
                                                                  self.cpu, self.ram, self.hv, self.ev))
        
class Vm:
    def __init__(self, name, add_day, cpu, ram, ID, days, user_price):
        self.name = name
        self.add_day = add_day
        self.cpu = cpu
        self.ram = ram
        self.ID = ID
        self.days = days
        self.user_price = user_price
        self.predict_hv = 0
        self.predict_ev = 0
        self.predict_price = 0
        
    def debug(self):
        print("name: {}, cpu: {}, ram: {}, days: {}, user_price: {}, pct_hv: {}, pct_ev: {}, pct_price: {}".format(
            self.name, self.cpu, self.ram, self.days, self.user_price, 
            self.predict_hv, self.predict_ev, self.predict_price))

class XJBG:
    def __init__(self):
        self.m_servers = []
        self.m_requests = []
        self.t = 0
        self.model1 = linear_model.LinearRegression(fit_intercept=False,normalize=False)
        self.model2 = linear_model.LinearRegression(fit_intercept=False,normalize=False)
    
    def read_data(self, path):   
        with open(path, 'r') as f:
            n = int(f.readline())
            for i in range(n):
                line = f.readline().strip()[1:-1].split(sep=', ')
                s = list(map(int, line[1:]))
                server = Server(line[0], s[0], s[1], s[2], s[3])
                self.m_servers.append(server)

            m = int(f.readline())
            dt = {}
            for i in range(m):
                line = f.readline().strip()[1:-1].split(sep=', ')
                s = list(map(int, line[1:]))
                dt[line[0]]= (s[0], s[1])

            t = int(f.readline().strip().split(sep=' ')[0])
            self.t = t
            for i in range(t):
                r = int(f.readline())
                reqs = []
                for j in range(r):
                    line = f.readline().strip()[1:-1].split(sep=', ')
                    if len(line) == 5:
                        req = Vm(line[1], i, dt[line[1]][0], dt[line[1]][1], 
                                 int(line[2]), int(line[3]), int(line[4]))
                        self.m_requests.append(req)
                        
    def debug(self):
        for svr in self.m_servers:
            svr.debug()
        for req in self.m_requests:
            req.debug()
            
    def cal(self, cpu, ram):
        return [cpu, ram]
    
    def train(self):
        X1, y1, X2, y2 = [], [], [], []
        for svr in self.m_servers:
            X1.append([svr.cpu, svr.ram])
            y1.append([svr.hv, svr.ev])
        self.model1.fit(X1, y1)
        print(self.model1.coef_, self.model1.intercept_, self.model1.score(X1, y1))
        
        
    
    def test(self):
        df = []
        for i in range(len(self.m_requests)):
            req = self.m_requests[i]
            result = self.model1.predict([[req.cpu, req.ram]])[0]
            y1, y2 = int(result[0]), int(result[1])
#             y1 = int(y1 * self.t / (self.t - req.add_day))
            y1 = req.user_price - y2 * req.days
            y1 *= 0.8
            
            req.predict_hv, req.predict_ev = y1, y2
            req.predict_price = y1 + y2 * req.days
            discount = float("%.3f" % (req.predict_price / req.user_price))
            
            df.append([req.name, req.ID, req.cpu, req.ram, req.add_day, req.days, req.predict_hv, 
                       req.predict_ev, req.predict_price, req.user_price, discount])
        df = pd.DataFrame(df)
        df.columns = ['name', 'id', 'cpu', 'ram', 'add_day', 'days', 'hv', 'ev', 'pct_price', 'user_price', 'discount']
        return df
    
    
xjbg = XJBG()
xjbg.read_data("./training-1.txt")
xjbg.train()
df = xjbg.test()
```

```python
import math
def sigmod(x):
    return 1 / (1 + math.exp(-x))

def yang(x):
    low_bound = 0.2
    m_tol_day = 800
    up_bound = 1
    x_split = m_tol_day * 0.7
    y_split = up_bound - (up_bound - low_bound) * 0.7

    a1 = (up_bound * up_bound - y_split * y_split) / (0 - x_split * x_split)
    b1 = up_bound * up_bound - a1 * 0
    a2 = (y_split * y_split - low_bound * low_bound) / (x_split * x_split - m_tol_day * m_tol_day)
    b2 = y_split * y_split - a2 * x_split * x_split

    if x < x_split:
        return a1 * x * x + b1
    else:
        return a2 * x * x + b2

xx = [i for i in range(800)]
l, r = [-5, 2]
# unit = tol / (r - l)


# y_tian = [1-(sigmod((x-300)/(len(xx)/10)) * 0.8) for x in xx]
y_yang = [yang(x) for x in xx]

# plt.plot(xx, y_tian, label='tian')
# plt.plot(xx, y_hao, label='hao')
plt.plot(xx, y_yang, label='yang')
plt.legend()
plt.show()
```
